```
# импорт всех необходимых библиотек и функций
from scipy.integrate import nquad
from numpy import exp, log
import numpy as np
from matplotlib import pyplot as plt
```

```
# функция вычисления константы в зависимости от входных параметров
def Const(D, wl, lam, h, w):
    return D / (16 * lam * h * log(4 * D / wl - 2 + 7 * wl / 2 * D))

# общая часть интеграла для всех функций
# параметры это константы
# а так же экспоненциальная функция и пределы интеграла
def Integral(D, wl, lam, h, w, y0, x0, exp_funk, borders):
    f = lambda x,y: exp_funk(x) * ((x0 - x) / ((y0 - y) ** 2 + (x0 - x) ** 2))
    f, err = nquad(f, borders)
    return f
```

```
# определяем константы
D, wj, lam, h, w, y0 = \
    2 * 10 ** -3, 8 * 10 ** -4, 50 * 10 ** -9, 10 * 10 ** -9, 30 * 10 ** -6, 10 * 10 ** -9

# определяем вектор x
x0 = np.arange(14.75, 15.25, 0.01) * 10 ** -6
```
